```python
import pandas as pd
import glob
```

```python
path = r'D:\JupyterNotebook\Dataset\click' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

click = pd.concat(li, axis=0, ignore_index=True)
```

```python
path = r'D:\JupyterNotebook\Dataset\clickAnnotated' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

clickAnnotated = pd.concat(li, axis=0, ignore_index=True)
```
